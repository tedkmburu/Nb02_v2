```python
for x in range(numberOfFiles):
    largestCurrent = np.amax(I[x])
    maxCurrentThreshhold.append(largestCurrent * 0.80)
    indexOfFirstLargeCurrent = np.argmax(I[x] > maxCurrentThreshhold[x])

#     print(indexOfFirstLargeCurrent)
#     print(I[indexOfFirstLargeCurrent])
#     print(maxCurrentThreshhold)

    t[x] = t[x][indexOfFirstLargeCurrent:]
    I[x] = I[x][indexOfFirstLargeCurrent:]
    V[x] = V[x][indexOfFirstLargeCurrent:]

    secondDerivativeCurrent.append(np.gradient(np.gradient(I[x])))
```

```python
allZeroIndexes = [ [0]*3 for i in range(numberOfFiles)]
zeroIndexes = [ [0]*3 for i in range(numberOfFiles)]

allMaxIndexes = [ [0]*3 for i in range(numberOfFiles)]
maxIndexes = [ [0]*3 for i in range(numberOfFiles)]
newMaxIndexes = [ [0]*3 for i in range(numberOfFiles)]

allIndexes = [ [0]*3 for i in range(numberOfFiles)]

posThreshValues = [ [0]*3 for i in range(numberOfFiles)]
negThreshValues = [ [0]*3 for i in range(numberOfFiles)]
posHoldValues = [ [0]*3 for i in range(numberOfFiles)]
negHoldValues = [ [0]*3 for i in range(numberOfFiles)]
```

```python
def isIncreasing(L):
    return L[0] < L[len(L) - 1]
def isDecreasing(L):
    return L[0] > L[len(L) - 1]
def non_increasing(L):
    return all(x>=y for x, y in zip(L, L[1:]))
def non_decreasing(L):
    return all(x<=y for x, y in zip(L, L[1:]))
def monotonic(L):
    return non_increasing(L) or non_decreasing(L)

# 0 to + is thresh
# + to 0 is hold
# 0 to - thresh on neg
# - to 0 hold on neg

# this will sort runs into the piles and make all the currents positive

def sortIntoSide(time, voltage, current, secondDerivative, q):
    if isIncreasing(voltage) and isIncreasing(current):
        current = [ abs(x) for x in current]
        posThreshValues[q].append([time, voltage, current, secondDerivative])
    elif isDecreasing(voltage) and isDecreasing(current):
        current = [ abs(x) for x in current]
        posHoldValues[q].append([time, voltage, current, secondDerivative])
    elif isDecreasing(voltage) and isIncreasing(current):
        current = [ abs(x) for x in current]
        negThreshValues[q].append([time, voltage, current, secondDerivative])
    elif isIncreasing(voltage) and isDecreasing(current):
        current = [ abs(x) for x in current]
        negHoldValues[q].append([time, voltage, current, secondDerivative])
    else: 
        print("???")
```

```python
for q in range(numberOfFiles):
    # indexes of values that are pretty much 0
    zeroThreshhold = 1e-5
    allZeroIndexes[q] = []
    for index, x in enumerate(I[q]):
        if x < zeroThreshhold:
            allZeroIndexes[q].append(index)

    # find the zero values that begin or end a run
    zeroIndexes[q].append(allZeroIndexes[q][0])
    for x in range(len(allZeroIndexes[q]) - 1):
        if allZeroIndexes[q][x]+1 != allZeroIndexes[q][x+1]:
            zeroIndexes[q].append(allZeroIndexes[q][x])
            zeroIndexes[q].append(allZeroIndexes[q][x+1])
```

```python
for q in range(numberOfFiles):
    # find all the values that hit the max
    for index, x in enumerate(I[q]):
        if x > maxCurrentThreshhold[q]:
            allMaxIndexes[q].append(index)

    # find the max values that start or end a run
    for x in range(len(allMaxIndexes[q]) - 1):
        if allMaxIndexes[q][x]+1 != allMaxIndexes[q][x+1]:
            maxIndexes[q].append(allMaxIndexes[q][x])
            maxIndexes[q].append(allMaxIndexes[q][x+1])

    for x in range(len(allMaxIndexes[q]) - 1):
        if allMaxIndexes[q][x]+1 != allMaxIndexes[q][x+1]:
            newMaxIndexes[q].append(allMaxIndexes[q][x]+1)
            newMaxIndexes[q].append(allMaxIndexes[q][x+1]+1)
```

```python
for q in range(numberOfFiles):
    # combine all zero and max indexes into one sorted array
    allIndexes[q] = [*zeroIndexes[q], *maxIndexes[q]]
    allIndexes[q] = sorted(allIndexes[q])

    for x in range(3, len(allIndexes[q]) - 1, 4):
        allIndexes[q][x] = allIndexes[q][x] + 1

    goodRuns = []
    for x in range(0, len(allIndexes[q]) - 4, 4):
        if monotonic( V[q][allIndexes[q][x]:allIndexes[q][x+3]] ):
            goodRuns.append(allIndexes[q][x])
            goodRuns.append(allIndexes[q][x+1])
            goodRuns.append(allIndexes[q][x+2])
            goodRuns.append(allIndexes[q][x+3])

    allIndexes[q] = np.copy(goodRuns)
#     print(q)
```

```python
for q in range(numberOfFiles):
    # loop through each run and put the V and I values into an array
    # then make an array with all the arrays
    # the indexes in the original current array is also passed
    timesToLoop = len(zeroIndexes[q]) - 1
    # timesToLoop = 20
    for x in range(0, timesToLoop, 2):
        voltageRun = []
        currentRun = []
        timeRun = []
        secondDerivativeRun = []
        start = allIndexes[q][x]
        stop = allIndexes[q][x+1]
        for y in range(start, stop):
            voltageRun.append(V[q][y])
            currentRun.append(I[q][y])
            timeRun.append(t[q][y])
            secondDerivativeRun.append(I[q][y])
        if len(currentRun) > 4:
            sortIntoSide(timeRun, voltageRun, currentRun, np.gradient(np.gradient(currentRun)), q)
```

```python
# take second derivative of each current array
posThreshValuesGradients = [ [0]*3 for i in range(numberOfFiles)]
negThreshValuesGradients = [ [0]*3 for i in range(numberOfFiles)]
posHoldValuesGradients = [ [0]*3 for i in range(numberOfFiles)]
negHoldValuesGradients = [ [0]*3 for i in range(numberOfFiles)]

posThreshValuesLargestGradient = [ [0]*3 for i in range(numberOfFiles)]
negThreshValuesLargestGradient = [ [0]*3 for i in range(numberOfFiles)]
posHoldValuesLargestGradient = [ [0]*3 for i in range(numberOfFiles)]
negHoldValuesLargestGradient = [ [0]*3 for i in range(numberOfFiles)]

print(posThreshValues)

for q in range(numberOfFiles):
    for x in posThreshValues[q]:
        posThreshValuesGradients[q].append(x[3])
    for x in negThreshValues[q]:
        negThreshValuesGradients[q].append(x[3])
    for x in posHoldValues[q]:
        posHoldValuesGradients[q].append(x[3])
    for x in negHoldValues[q]:
        negHoldValuesGradients[q].append(x[3])

    # find the index of the largest value in the 2nd derivative


    # 0 to + is thresh
    # + to 0 is hold
    # 0 to - thresh on neg
    # - to 0 hold on neg

    for x in posThreshValuesGradients[q]:
        largestValue = np.argmax(np.absolute(x[1:len(x) - 1])) + 1
        posThreshValuesLargestGradient[q].append(largestValue)
    for x in negThreshValuesGradients[q]:
        largestValue = np.argmax(np.absolute(x[1:len(x) - 1])) + 1
        negThreshValuesLargestGradient[q].append(largestValue)
    for x in posHoldValuesGradients[q]:
        largestValue = np.argmax(np.absolute(x[1:len(x) - 1])) + 1
        posHoldValuesLargestGradient[q].append(largestValue)
    for x in negHoldValuesGradients[q]:
        largestValue = np.argmax(np.absolute(x[1:len(x) - 1])) + 1
        negHoldValuesLargestGradient[q].append(largestValue)
```

```python
# get the voltage when the largest second derivative happens
posThreshValuesVoltage = []
negThreshValuesVoltage = []
posHoldValuesVoltage = []
negHoldValuesVoltage = []

for x in range(len(posThreshValuesGradients)):
    voltage = posThreshValues[x][1][posThreshValuesLargestGradient[x]]
    posThreshValuesVoltage.append(voltage)
for x in range(len(negThreshValuesGradients)):
    voltage = negThreshValues[x][1][negThreshValuesLargestGradient[x]]
    negThreshValuesVoltage.append(voltage)
for x in range(len(posHoldValuesGradients)):
    voltage = posHoldValues[x][1][posHoldValuesLargestGradient[x]]
    posHoldValuesVoltage.append(voltage)
for x in range(len(negHoldValuesGradients)):
    voltage = negHoldValues[x][1][negHoldValuesLargestGradient[x]]
    negHoldValuesVoltage.append(voltage)
```

```python
posThreshValuesCurrent = []
negThreshValuesCurrent = []
posHoldValuesCurrent = []
negHoldValuesCurrent = []

for x in range(len(posThreshValuesGradients)):
    current = posThreshValues[x][2][posThreshValuesLargestGradient[x]]
    posThreshValuesCurrent.append(current)
for x in range(len(negThreshValuesGradients)):
    current = negThreshValues[x][2][negThreshValuesLargestGradient[x]]
    negThreshValuesCurrent.append(current)
for x in range(len(posHoldValuesGradients)):
    current = posHoldValues[x][2][posHoldValuesLargestGradient[x]]
    posHoldValuesCurrent.append(current)
for x in range(len(negHoldValuesGradients)):
    current = negHoldValues[x][2][negHoldValuesLargestGradient[x]]
    negHoldValuesCurrent.append(current)
```

```python
def average(array):
    return sum(array) / len(array)

print("Positive Threshold Voltage: ", average(posThreshValuesVoltage), "Standard Deviation: ", np.std(posThreshValuesVoltage))
print("Negative Threshold Voltage: ", average(negThreshValuesVoltage), "Standard Deviation: ", np.std(negThreshValuesVoltage))
print("Positive Hold Voltage: ", average(posHoldValuesVoltage), "Standard Deviation: ", np.std(posHoldValuesVoltage))
print("Negative Hold Voltage: ", average(negHoldValuesVoltage), "Standard Deviation: ", np.std(negHoldValuesVoltage))
```
